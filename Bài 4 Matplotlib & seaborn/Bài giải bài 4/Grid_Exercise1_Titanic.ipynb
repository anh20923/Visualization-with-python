## Grid_Exercise1_Titanic
TVào ngày 15 tháng 4 năm 1912, trong chuyến hải trình đầu tiên của mình, tàu Titanic đã chìm sau khi va chạm với một tảng băng trôi, đã có 1502 nạn nhân mãi mãi ra đi trong tổng số 2224 hành khách và thủy thủ đoàn. 

Tập dữ liệu **train.csv** chứa thông tin về hành khách trên tàu Titanic với ý nghĩa các cột như sau:
1.	PassengerId: ID của mỗi hành khách.
2.	Survived: cho biết hành khách còn sống sót (1) hay không (0).
3.	Pclass: Hạng vé của hành khách (1 = Hạng nhất, 2 = Hạng hai, 3 = Hạng ba).
4.	Name: Tên đầy đủ của hành khách.
5.	Sex: Giới tính của hành khách.
6.	Age: Tuổi của hành khách.
7.	SibSp: Số lượng anh chị em hoặc vợ/chồng đi cùng trên tàu.
8.	Parch: Số lượng cha mẹ hoặc con cái đi cùng trên tàu.
9.	Ticket: Mã số vé.
10.	Fare: Giá vé.
11.	Cabin: Số hiệu phòng ngủ trên tàu.
12.	Embarked: Cảng mà hành khách lên tàu (C = Cherbourg; Q = Queenstown; S = Southampton).

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Câu 1: đọc dữ liệu train.csv vào biến titanic
# Đọc tập tin dữ liệu, dữ liệu về các hành khách trên tàu Titanic
titanic = pd.read_csv('train.csv')
titanic.head()
```

```python
# Câu 2: Vẽ biểu đồ boxplot trên lưới FacetGrid để thể hiện phân bố dữ liệu của giá vé theo giới tính
# chia cột theo Pclass
g = sns.FacetGrid(data=titanic,col='Pclass')
g.map(sns.boxplot,'Sex','Fare',order=None)
```

```python
# Câu 3: Vẽ biểu đồ countplot trên lưới catplot để thể hiện số hành khách theo Survived và Embarked (nơi lên tàu)
# chia cột theo Pclass
sns.catplot(data=titanic,col='Pclass',kind='count',x='Embarked',hue='Survived')
```

```python
# Câu 4: Vẽ biểu đồ countplot trên lưới catplot để Thống kê số lượng hành khách theo Pclass, Sex và Survived
# Chia cột theo Sex, chia màu theo Survived và x là Pclass
sns.catplot(data=titanic,col='Sex',kind='count',x='Pclass',hue='Survived')
```

```python
# Câu 5: Vẽ biểu đồ displot thể hiện phân phối dữ liệu tuổi của hành khách
# Chia cột theo Sex, chia màu theo Survived và có hiển thị kde
sns.displot(data=titanic,x='Age',col='Sex',hue='Survived',kind='kde')
```

```python
# Câu 6: Sử dụng lưới PairGrid để thể hiện phân phối dữ liệu (histplot) của Age, Fare 
# và tương quan (scatterplot) giữa Age và Fare, phân màu theo Pclass
g = sns.PairGrid(titanic[['Age','Fare','Pclass']],hue='Pclass',palette='Set1')
g.map_diag(sns.histplot)
g.map_offdiag(sns.scatterplot)
g.add_legend()
```

```python
# Câu 7) Sử dụng PairPlot để phân tích phân phối tuổi, giá vé của các hành khách theo Survived
sns.pairplot(titanic[['Age','Fare','Survived']],hue='Survived')
```

```python
# Câu 8) Sử dụng JointGrid để phân tích phân phối và tương quan tuổi, giá vé của các hành khách theo Survived
g = sns.JointGrid(data=titanic,x='Age',y='Fare',hue='Survived')
g.plot(sns.scatterplot,sns.kdeplot)
```

```python
# Câu 9: Subplots để tạo một lưới của các đồ thị sử dụng Matplotlib
fig,ax = plt.subplots(2,2,figsize=(10,8))
sns.kdeplot(data=titanic,x='Age',ax=ax[0,0],hue='Sex')
sns.countplot(data=titanic,x='Sex',ax=ax[0,1],hue='Sex')
sns.scatterplot(data=titanic,x='Age',y='Fare',ax=ax[1,0],hue='Sex')
plt.tight_layout()
```

```python
# Câu 10: Vẽ countplot để thể hiện sự phân bố của 'SibSp' và 'Parch' trên cùng figure bằng subplot
plt.figure(figsize=(10,5))

plt.subplot(1,2,1)
sns.countplot(data=titanic,x='SibSp',hue='SibSp',palette='Set1')

plt.subplot(1,2,2)
sns.countplot(data=titanic,x='Parch',hue='Parch',palette='Set1')
```
