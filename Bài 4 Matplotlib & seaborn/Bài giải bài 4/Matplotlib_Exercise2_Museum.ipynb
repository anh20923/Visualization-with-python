Tập dữ liệu **Museum_visitors.csv** chứa thông tin về tổng số lượng khách tham quan mỗi tháng từ 1/2014 đến 12/2019 các bảo tàng tại Los Angeles, California, Mỹ; bao gồm:

3.	**Chinese American Museum**: tập trung vào việc bảo tồn và truyền đạt lịch sử, nghệ thuật và văn hóa của cộng đồng người Mỹ gốc Trung Quốc. Bảo tàng này mang đến cái nhìn đa chiều về sự đóng góp và trải nghiệm của cộng đồng này trong lịch sử của Los Angeles và Hoa Kỳ.
4.	**America Tropical Interpretive Center**: nơi bảo tồn tác phẩm nghệ thuật tường của nghệ sĩ Diego Rivera, được biết đến với tên gọi "America Tropical." Bức tranh này tập trung vào chủ đề lịch sử và chính trị và đã trở thành biểu tượng nghệ thuật chính trị tại Los Angeles.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Câu 1: 
# a) Đọc dữ liệu từ tập tin Museum_visitors.csv và lưu vào biến museum.  
museum = pd.read_csv('Museum_visitors.csv')
# Hiển thị 5 dòng dữ liệu đầu của museum
museum.head()
```

```python
# Hiển thị 5 dòng dữ liệu cuối của museum
museum.tail()
```

```python
# Câu 2: chuyển cột Month sang kiểu datetime, tạo cột month (tháng - kiểu số) và year (năm)
museum.Month = pd.to_datetime(museum.Month,format="%y-%b")
museum.head()
```

```python
museum['month'] = museum.Month.dt.month
museum['year'] = museum.Month.dt.year
museum.head()
```

```python
# Lưu ý sau khi chuyển sang datetime, cột Month hiển thị dạng %Y-%m-%d nhưng cần hiểu đây là dữ liệu theo tháng
# tức là tổng số khách tham quan trong 1 tháng; không phải trong 1 ngày
```

```python
museum.year.unique()
```

```python
# Câu 3: Vẽ biểu đồ line thể hiện Lượng khách tham quan bảo tàng Avila Adobe từ năm 2014-2019
plt.figure(figsize=(12,5))
plt.plot(museum.Month,museum['Avila Adobe'])
```

```python
museum.columns
```

```python
# Câu 4: Vẽ biểu đồ line thể hiện Số lượng khách tham quan các bảo tàng
# Đặt kích thước biểu đồ
plt.figure(figsize=(12,6))

# Vẽ biểu đồ line cho từng bảo tàng và đặt nhãn
plt.plot(museum.Month,museum['America Tropical Interpretive Center'],label='America Tropical Interpretive Center')
plt.plot(museum.Month,museum['Avila Adobe'],label='Avila Adobe')
plt.plot(museum.Month,museum['Chinese American Museum'],label='Chinese American Museum')
plt.plot(museum.Month,museum['Firehouse Museum'],label='Firehouse Museum')
# Đặt tiêu đề và nhãn
plt.title('Số lượng khách tham quan các bảo tàng')

# Hiển thị legend, lưới
plt.legend()
plt.grid(True)
# Hiển thị biểu đồ
plt.show()
```

```python
# Câu 5: # Vẽ biểu đồ lượng du khách tham quan Bảo tàng Avila Adobe và America Tropical Interpretive Center trung bình theo tháng
# Tính lượng du khách tham quan các bảo tàng trung bình theo tháng
df = museum.groupby(['Month','month'],as_index=False)[['America Tropical Interpretive Center', 'Avila Adobe']].mean()
df.head()
```

```python
# Vẽ biểu đồ 
plt.plot(df.Month,df['America Tropical Interpretive Center'],label='America Tropical Interpretive Center')
plt.plot(df.Month,df['Avila Adobe'],label='Avila Adobe')
```

```python
# Câu 6: Vẽ biểu đồ Tổng lượng du khách tham quan các Bảo tàng theo năm
# Tính tổng lượng khách tham quan các bảo tàng mỗi năm
df = museum.groupby('year',as_index=False)[['America Tropical Interpretive Center', 'Avila Adobe','Chinese American Museum', 'Firehouse Museum']].sum()
df.head()
```

```python
# Vẽ biểu đồ 
plt.plot(df.year,df['America Tropical Interpretive Center'],label='America Tropical Interpretive Center')
plt.plot(df.year,df['Avila Adobe'],label='Avila Adobe')
plt.plot(df.year,df['Chinese American Museum'],label='Chinese American Museum')
plt.plot(df.year,df['Firehouse Museum'],label='Firehouse Museum')
```

```python
# Câu 7: Vẽ biểu đồ barh thể hiện trung bình lượng du khách tham quan các bảo tàng mỗi năm
# Tính trung bình lượng du khách tham quan các bảo tàng mỗi năm
df = museum.groupby('year',as_index=False)[['America Tropical Interpretive Center', 'Avila Adobe','Chinese American Museum', 'Firehouse Museum']].mean()
df.head()
```

```python
# Vẽ biểu đồ 
plt.barh(df.year,width=df['America Tropical Interpretive Center'])
```
