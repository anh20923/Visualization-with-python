```python
# Câu 5a) Hiển thị 10 quốc gia có tổng số huy chương lớn nhất, tách theo giới tính 
# Tìm top_10
idx = athletes.groupby('nationality')['total'].sum().sort_values(ascending=False).head(10).index
top_10 = athletes[athletes['nationality'].isin(idx)]
top_10
```

```python
# Vẽ biểu đồ
plt.figure(figsize=(12,6))
sns.barplot(data=top_10, x='nationality', y='total', estimator=sum, errorbar=None, hue='sex',order=idx)
plt.title('10 quốc gia có nhiều huy chương nhất')
plt.xlabel('Quốc gia')
plt.ylabel('Số huy chương')
plt.show()
```

```python
# b) Vẽ biểu đồ stacked bar chart thể hiện số huy chương vàng, bạc, đồng của 10 quốc gia có tổng số huy chương cao nhất
# Tính số huy chương vàng, bạc, đồng của 10 quốc gia có tổng số huy chương cao nhất
top_10_stat = top_10.groupby('nationality')\
        [['gold','silver','bronze', 'total']].sum().sort_values(by='total', ascending=False)
top_10_stat
```

```python
# Vẽ biểu đồ stacked bar chart 
plt.figure(figsize=(10,7))

sns.barplot(top_10_stat, x = top_10_stat.index, y = 'gold', label='Vàng', color = 'gold')
sns.barplot(top_10_stat, x = top_10_stat.index, y = 'silver', bottom=top_10_stat['gold'], label='Bạc', color = 'silver')
sns.barplot(top_10_stat, x = top_10_stat.index, y = 'bronze', bottom=top_10_stat['gold']+top_10_stat['silver'], label='Đồng', color = 'peru')

plt.legend(loc='upper center', title='Huy chương')
plt.title('Thống kê số huy chương vàng, bạc, đồng của 10 quốc gia có nhiều huy chương nhất')
plt.ylabel('Số huy chương')
plt.show()
```

```python
# Câu 6a: Tạo dataframe basket_row gồm các VĐV thi đấu bóng rổ và đua thuyền
basket_row = athletes[athletes.sport.isin(['basketball', 'rowing'])]
basket_row.head()
```

```python
# b) Biểu đồ thể hiện Cân nặng trung bình của VĐV Đua thuyền vs Bóng rổ.
# Bạn có nhận xét gì từ biểu đồ này?
sns.barplot(basket_row, x = 'sport', y = 'weight', errorbar=None)
plt.bar_label()
plt.title('Cân nặng trung bình của VĐV Đua thuyền vs Bóng rổ')
plt.show()
```

```python
# c) Biểu đồ phân phối số VĐV theo cân nặng giữa Basketball và Rowing. Cho nhận xét.
sns.displot(basket_row, hue = 'sport', x = 'weight')
plt.title('Biểu đồ phân phối cân nặng của các VĐV Đua thuyền vs Bóng rổ')
plt.ylabel('Số VĐV')
plt.show()
```

```python
# d) Biểu đồ thể hiện Chiều cao trung bình của VĐV Đua thuyền vs Bóng rổ.
# Bạn có nhận xét gì từ biểu đồ này?
sns.barplot(basket_row, x = 'sport', y = 'height', errorbar=None)
plt.title('Chiều cao trung bình của VĐV Đua thuyền vs Bóng rổ')
plt.show()
```

```python
# e) Biểu đồ phân phối chiều cao của VĐV Đua thuyền vs Bóng rổ. Cho nhận xét.
sns.displot(basket_row, hue = 'sport', x = 'height')
plt.title('Biểu đồ phân phối chiều cao của VĐV Đua thuyền vs Bóng rổ')
plt.ylabel('Số VĐV')
plt.show()
```

```python
# f) Biểu đồ boxplot thể hiện phân bố chiều cao của các VĐV Đua thuyền vs Bóng rổ
sns.boxplot(data=basket_row, x='sport', y='height');
```

```python
# g) Biểu đồ violinplot thể hiện phân bố chiều cao của các VĐV Đua thuyền vs Bóng rổ, phân màu theo giới tính
sns.violinplot(data=basket_row, x='sport', y='height', hue = 'sex');
```

```python
# h) Vẽ scatterplot thể hiện mối quan hệ giữa chiều cao và cân nặng của các VĐV bóng rổ, đua thuyền
# phân màu theo môn thi
sns.scatterplot(basket_row, x = 'height',  y = 'weight', hue = 'sport')
plt.title('Mối tương quan giữa chiều cao và cân nặng của các VĐV bóng rổ, đua thuyền')
plt.xlabel('Chiều cao')
plt.ylabel('Cân nặng')
plt.show()
```

```python
# i) Vẽ regressionplot cho chiều cao và cân nặng của VĐV bóng rổ
df_basketball = athletes[athletes.sport == 'basketball']
plt.figure(figsize=(8,5))
sns.regplot(data=df_basketball, x='height', y='weight')
plt.title('Biểu đồ hồi quy tuyến tính cho chiều cao và cân nặng của VĐV bóng rổ')
plt.ylabel('Cân nặng')
plt.xlabel('Chiều cao')
plt.show()
```

```python
# k) Vẽ residualplot cho chiều cao và cân nặng của VĐV bóng rổ
plt.figure(figsize=(8,5))
sns.residplot(data=df_basketball, x='height', y='weight')
plt.title('Biểu đồ phần dư cho chiều cao và cân nặng của VĐV bóng rổ')
plt.ylabel('Cân nặng')
plt.xlabel('Chiều cao')
plt.show()
```

```python
# l) Hàm mật độ xác suất chiều cao của các VĐV bóng rổ theo giới tính
sns.displot(data=df_basketball, x='height', hue='sex', kind = 'kde')
plt.title('Hàm mật độ xác suất chiều cao của các VĐV bóng rổ theo giới tính', y=1.03)
plt.show()
```

```python
# Câu 7) Vẽ heatmap thể hiện Số huy chương vàng/bạc/đồng của 10 quốc gia có nhiều huy chương nhất
plt.figure(figsize=(15,9))
sns.heatmap(top_10_stat[['gold', 'silver', 'bronze']], cbar=True, cmap='YlGnBu', linewidths=0.3, annot=True, fmt='g')
plt.title('Số huy chương vàng/bạc/đồng của 10 quốc gia có nhiều huy chương nhất')
plt.ylabel('Quốc gia')
plt.show()
```
