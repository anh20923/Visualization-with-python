```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

# Sử dụng FacetGrid của Seaborn

#### Dữ liệu tips có sắn trong bộ dataset của Seaborn

```python
tips = sns.load_dataset("tips")
tips.head()
```

```python
tips.sex.unique()
```

```python
g = sns.FacetGrid(tips, col="sex")
```

```python
#Khởi tạo grid, tách tập dữ liệu theo "time"
g = sns.FacetGrid(tips, col="time",hue='sex')

#Sử dụng hàm map() để vẽ đồ thị trên các ô trong lưới
g.map(sns.kdeplot, "total_bill")
g.add_legend()
plt.show()
```

```python
#Khởi tạo grid, tách tập dữ liệu theo "time"
g = sns.FacetGrid(tips, col="time")

#Sử dụng hàm map() để vẽ đồ thị trên các ô dữ liệu trong lưới
g.map(sns.barplot,"day","total_bill",ci=0)
```

```python
g = sns.FacetGrid(tips, col="day", col_wrap=3)

#Sử dụng hàm map() để vẽ đồ thị trên các ô dữ liệu trong lưới
g.map(sns.scatterplot, "total_bill","tip")
```

```python
g = sns.FacetGrid(tips, col="time", hue="sex")

#Sử dụng hàm map() để vẽ đồ thị trên các ô dữ liệu trong lưới
g.map(sns.scatterplot, "total_bill","tip")
g.add_legend()
```

```python
g = sns.FacetGrid(tips, col="time",row='day',hue="sex") #(4,2)

#Sử dụng hàm map() để vẽ đồ thị trên các ô dữ liệu trong lưới
g.map(sns.scatterplot, "total_bill","tip")
g.add_legend()
```

```python
sns.displot(data=tips,x='total_bill',col='day',row='time',kind='kde',hue='sex')
```

```python
sns.scatterplot(data=tips,x='total_bill',y='tip',hue='sex')
```

```python
sns.relplot(data=tips,x='total_bill',y='tip',col='time',hue='sex')
```

```python
sns.regplot(data=tips,x='total_bill',y='tip')
```

```python
sns.lmplot(data=tips,x='total_bill',y='tip',col='time',hue='sex')
```

```python
g = sns.FacetGrid(tips,col='time')

#Sử dụng hàm map() để vẽ đồ thị trên các ô dữ liệu trong lưới
g.map(sns.boxplot,"sex", "total_bill")
```

```python
sns.catplot(data=tips, col="time",hue='sex',kind='box',y='total_bill',x='sex')
```

```python
sns.catplot(data=tips, col="time",kind='count',x='sex',hue='sex')
```

### Sử dụng PairGrid để phân tích các biến số trong tập dữ liệu

```python
g = sns.PairGrid(tips[['total_bill','tip','sex']],hue="sex")
```

```python
#Định nghĩa lưới
g = sns.PairGrid(tips[['total_bill','tip','sex']],hue="sex")

#Vẽ biểu đồ ở đường chéo trên, đường chéo dưới và trên đường chéo
g.map_upper(sns.scatterplot)
g.map_lower(sns.kdeplot)
g.map_diag(sns.kdeplot)
g.add_legend()
```

```python
#Định nghĩa lưới
g = sns.PairGrid(tips[['total_bill','tip','sex','size']],hue="sex")

#Vẽ biểu đồ ở đường chéo và khu vực không phải đường chéo 
g.map_upper(sns.scatterplot)
g.map_diag(sns.histplot)
g.map_lower(sns.regplot)
g.add_legend()
```

### Sử dụng Pairplot để phân tích nhanh các biến số trong tập dữ liệu

```python
sns.pairplot(tips[['total_bill','tip','sex']], hue="sex")
```

```python
sns.pairplot(tips[['total_bill','tip','sex']], hue="sex",kind='reg')
```

```python
#Vẽ lại biểu đồ ở trên bằng cách hiệu chỉnh các tham số của pairplot và map
g=sns.pairplot(tips[['total_bill','tip','sex']], hue="sex", diag_kind="hist")
g.map_lower(sns.kdeplot)
```

### Sử dụng JointGrid để phân tích cặp biến số

```python
#Định nghĩa lưới
g=sns.JointGrid(data=tips, x='total_bill', y='tip')

#Vẽ biểu đồ 
g.plot(sns.regplot, sns.kdeplot)
```

```python
#Định nghĩa lưới
g=sns.JointGrid(data=tips, x='total_bill', y='tip',hue="sex")

#Vẽ biểu đồ 
g.plot(sns.scatterplot, sns.boxplot)
```

### Sử dụng jointplot để phân tích cặp biến số

```python
sns.jointplot(x='total_bill', y='tip', data=tips)
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips, kind="reg")
```

## Sử dụng subplots của Matplotlib để linh động tùy chỉnh biểu đồ trên lưới

```python
# Khởi tạo subplots
fig, axes = plt.subplots(2, 2, figsize=(10, 8))

# Vẽ biểu đồ trên từng axes
sns.histplot(tips["total_bill"], ax=axes[0, 0])
sns.boxplot(x="day", y="total_bill", data=tips, ax=axes[0, 1])
sns.scatterplot(x="total_bill", y="tip", data=tips, ax=axes[1, 0])
axes[1, 0].set_title('bieu scatter')
sns.countplot(x="day", data=tips, ax=axes[1, 1])

# Tùy chỉnh layout và tiêu đề
plt.tight_layout()
plt.suptitle("Multiple Seaborn Plots", y = 1.05)

# Hiển thị figure
plt.show()
```

plt.tight_layout() là một hàm trong matplotlib được sử dụng để 
tự động điều chỉnh các khoảng cách giữa các thành phần trong hình vẽ 
để giảm thiểu việc chồng chéo, chồng lấn giữa các thành phần.

```python
# Khởi tạo figure
plt.figure(figsize=(8,8))

# Vẽ biểu đồ ở hàng 0
plt.subplot(2,2,1)
sns.histplot(tips["total_bill"])
plt.title('histogram')

plt.subplot(2,2,2)
sns.boxplot(x="day", y="total_bill", data=tips)
plt.title('boxplot')
# Vẽ biểu đồ scatter ở hàng 1
plt.subplot(2,2,(3,4))
sns.scatterplot(x="total_bill", y="tip", data=tips)
plt.title('scatterlot')

# Tùy chỉnh tiêu đề
plt.suptitle("Multiple Seaborn Plots")
plt.tight_layout()
# Hiển thị figure
plt.show()
```

```python
# Khởi tạo figure
plt.figure(figsize=(8,8))

# Vẽ biểu đồ ở hàng 0
plt.subplot(2,2,(1,3))
sns.histplot(tips["total_bill"])
plt.title('histogram')

plt.subplot(2,2,2)
sns.boxplot(x="day", y="total_bill", data=tips)
plt.title('boxplot')
# Vẽ biểu đồ scatter ở hàng 1
plt.subplot(2,2,4)
sns.scatterplot(x="total_bill", y="tip", data=tips)
plt.title('scatterlot')

# Tùy chỉnh tiêu đề
plt.suptitle("Multiple Seaborn Plots")
plt.tight_layout() 

# Hiển thị figure
plt.show()
```
