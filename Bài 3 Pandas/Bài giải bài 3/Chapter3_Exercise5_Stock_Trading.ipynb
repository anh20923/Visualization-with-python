```python
# Câu 2: Cho biết trong stocks1 có dữ liệu Null hay không?  
stocks1.isnull().any()
```

```python
stocks1.isnull().sum()
```

```python
stocks1[stocks1['high'].isnull()]
```

```python
# Nếu có, hãy thay thế với quy tắc sau:
# Nếu Null cột 'high' thì thay bằng giá trị max trên cột 'high' của mã chứng khoán đó
# Nếu Null cột 'low' thì thay bằng giá trị min trên cột 'low' của mã chứng khoán đó
stocks1['high'].fillna(stocks1.groupby(['symbol'])['high'].transform(max), inplace=True)
stocks1['low'].fillna(stocks1.groupby(['symbol'])['low'].transform(min), inplace=True)
```

```python
# Câu 3: Tạo dataframe stocks bằng cách gộp stocks1 và stocks2 theo dòng
stocks = pd.concat([stocks1, stocks2], ignore_index=True)
# Xem 15 dòng dữ liệu cuối của stocks
stocks.tail(15)
```

```python
# Câu 4: Tạo dataframe stocks_companies bằng cách gộp stocks và companies 
stocks_companies = stocks.merge(companies,left_on='symbol', right_on='name', how='inner')
# Xem 5 dòng dữ liệu đầu của stocks_companies
stocks_companies.head()
```

```python
# Câu 5: Cho biết giá (open, high, low, close) trung bình  và volume trung bình của mỗi công ty
cols = ['symbol','open','high','low','close','volume']
stocks_companies[cols].groupby('symbol').mean()
```

```python
# Câu 6: Cho biết giá đóng cửa (close) trung bình, lớn nhất và nhỏ nhất ở mỗi công ty
stocks_companies.groupby('symbol').close.agg(['mean', 'min', 'max'])
```

```python
# Câu 7: Tạo cột parsed_time trong stocks_companies bằng cách đổi thời gian sang định dạng DateTime
stocks_companies['parsed_time'] = pd.to_datetime(stocks_companies['date'])
# Cho biết kiểu dữ liệu của cột parsed_time
print(stocks_companies['parsed_time'].dtype)
# Hiển thị 5 dòng dữ liệu đầu của stocks_companies
stocks_companies.head()
```

```python
# Câu 8: Thêm cột result, nếu giá 'close' >  'open' thì cột result có giá trị 'up', ngược lại 'down'
stocks_companies.loc[stocks_companies['close'] > stocks_companies['open'], 'result'] = 'up'  
stocks_companies.loc[stocks_companies['close'] < stocks_companies['open'], 'result'] = 'down'  
stocks_companies.head()
```

```python
# Câu 9: Vẽ biểu đồ line chart thể hiện xu hướng khối lượng giao dịch của AMZN và nhận xét
AMZN = stocks_companies[stocks_companies.symbol == 'AMZN']
AMZN[['parsed_time', 'close']].plot(x = 'parsed_time', legend = False);
```
