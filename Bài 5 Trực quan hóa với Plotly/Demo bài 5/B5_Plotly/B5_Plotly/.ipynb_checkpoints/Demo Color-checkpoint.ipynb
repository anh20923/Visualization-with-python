```python
import plotly.express as px
import pandas as pd
```

```python
# Đọc dữ liệu từ tệp CSV
df = pd.read_csv('data/gapminder.csv')

# Tính giá trị trung bình của lượng CO2 phát thải
average_co2 = df['co2'].mean()

# Thêm cột mới vào DataFrame cho sự chênh lệch CO2 so với giá trị trung bình
df['co2_deviation'] = df['co2'] - average_co2

df.head()
```

 Quốc gia nào phát thải CO2 nhiều hơn mức trung bình toàn cầu có thể được xem là có tác động môi trường lớn hơn do hoạt động công nghiệp, giao thông, hay năng lượng tiêu thụ. Ngược lại, quốc gia phát thải thấp hơn mức trung bình có thể được coi là ít tác động hơn.

### Bar plot 

```python
df.info()
```

```python
df.groupby("region").mean(numeric_only=True).reset_index()
```

```python
# Tạo biểu đồ cột với thang màu rời rạc

fig = px.bar(df.groupby("region").mean(numeric_only=True).reset_index(), 
             x='region', y='co2_deviation',
             title="Trung bình chênh lệch lượng CO2 phát thải ở các châu lục",
             labels={'co2_deviation': 'Chênh lệch CO2 (tấn/người)'},
             color='region'
            )

# Hiển thị biểu đồ
fig.show()
```

```python
# Tạo biểu đồ cột với thang màu phân kỳ
fig = px.bar(df, x='region', y='co2_deviation',
             title="Sự chênh lệch lượng CO2 phát thải so với mức trung bình toàn cầu",
             labels={'co2_deviation': 'Chênh lệch CO2 (tấn/người)'},
             hover_name=df['country'],
             color='co2_deviation',
             color_continuous_scale=px.colors.diverging.PiYG
            )

# Hiển thị biểu đồ
fig.show()
```

```python
df.head()
```

```python
# Tạo stacked bar chart với thang màu phân kỳ
fig = px.bar(df, x='region', y='co2_deviation',
             title="Chênh lệch lượng CO2 phát thải so với mức trung bình toàn cầu",
             labels={'co2_deviation': 'Chênh lệch CO2 (tấn/người)'},
             hover_name=df['country'],
             color='co2_deviation',
             color_continuous_scale=px.colors.sequential.Reds
            )

# Hiển thị biểu đồ
fig.show()
```

```python
# Tạo stacked bar chart
fig = px.bar(df, x="region", y="population", color="income_level",
             title="Dân số - Mức Thu nhập của các quốc gia trong các châu lục",
             labels={"population": "Dân số", "income_level": "Mức Thu nhập"},
             hover_name=df['country'],
             color_discrete_sequence=px.colors.qualitative.Safe )

# Hiển thị biểu đồ
fig.show()
```

```python
# Vẽ density_heatmap
fig = px.density_heatmap(df, x="income", y="co2_deviation", nbinsx=30, nbinsy=30, color_continuous_scale=px.colors.sequential.Viridis)

# Hiển thị biểu đồ
fig.show()
```

```python
df.head()
```

```python
# Vẽ density_heatmap
fig = px.density_heatmap(df, x="life_exp", y="co2_deviation", nbinsx=30, nbinsy=30, color_continuous_scale="Reds")

# Hiển thị biểu đồ
fig.show()
```

```python
score = [4,7,9,6,8]
name = ['a','b','c','d','e']
```

```python
def color_score(s):
    lst_color = []
    for x in s:
        if x<5:
            lst_color.append('red')
        else:
            lst_color.append('blue')
    return lst_color
```

```python
colors = color_score(score)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.bar(name,score,color=colors)
```
