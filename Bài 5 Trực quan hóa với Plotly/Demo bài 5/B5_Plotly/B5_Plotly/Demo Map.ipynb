```python
import plotly.express as px
import pandas as pd
```

### Đọc hiểu dữ liệu

```python
# Tải tập dữ liệu Gapminder là dữ liệu năm 2017
df = pd.read_csv('data\gapminder.csv')

df.head()
```

### Trực quan hóa không gian địa lý

#### Choropleth map

```python
df1 = px.data.gapminder()
df1.head()
```

```python
# Choropleth map với dải màu tuần tự
fig = px.choropleth(data_frame=df1[df1.year == 2007],
                    locations='iso_alpha',
                    color="lifeExp", # Thuộc tính để thể hiện màu sắc
                     # Kiểu chiếu bản đồ
                    color_continuous_scale=px.colors.sequential.Peach,
                    title='Bản đồ tuổi thọ trung bình trên thế giới') # Tiêu đề bản đồ

fig.update_layout(
    width=1000,  # Chiều rộng của biểu đồ bằng pixel
    height=500,   # Chiều cao của biểu đồ bằng pixel
    margin = dict(t=40, l=50, r=25, b=25))

fig.show()
```

```python
# Choropleth map với dải màu tuần tự
fig = px.choropleth(data_frame=df,
                    locations='country',
                    locationmode='country names',
                    color="life_exp", # Thuộc tính để thể hiện màu sắc
                     # Kiểu chiếu bản đồ
                    title='Bản đồ tuổi thọ trung bình trên thế giới') # Tiêu đề bản đồ

fig.update_layout(
    width=1000,  # Chiều rộng của biểu đồ bằng pixel
    height=500,   # Chiều cao của biểu đồ bằng pixel
    margin = dict(t=40, l=50, r=25, b=25))

fig.show()
```

```python
# Choropleth map với dải màu tuần tự
fig = px.choropleth(df,
                    locations='country',
                    locationmode='country names',
                    color='life_exp',
                    hover_name='country',
                    color_continuous_scale=px.colors.sequential.Plasma,
                    projection='natural earth',
                    title='Bản đồ tuổi thọ trung bình trên thế giới') # Tiêu đề bản đồ

fig.update_layout(
    width=1000,  # Chiều rộng của biểu đồ bằng pixel
    height=500,   # Chiều cao của biểu đồ bằng pixel
    margin = dict(t=40, l=50, r=25, b=25))

fig.show()
```

```python

# Choropleth map với dải màu phân kỳ
fig = px.choropleth(df, 
                    locations="country", 
                    locationmode='country names',
                    color="income", 
                    hover_name="country", 
                    color_continuous_scale=px.colors.diverging.Spectral, #Dải màu phân kỳ
                    projection="natural earth",
                    title="Thu nhập của các quốc gia theo Choropleth Map")

fig.update_layout(
    width=1000,  # Chiều rộng của biểu đồ bằng pixel
    height=500,   # Chiều cao của biểu đồ bằng pixel
    margin = dict(t=40, l=50, r=25, b=25))
# Show the map
fig.show()
```

```python
# Choropleth map với màu rời rạc
fig = px.choropleth(df,
                    locations="country", # Tên quốc gia
                    locationmode='country names',
                    color="region", # Thuộc tính để thể hiện màu sắc
                    hover_name="country", # Tên quốc gia để hiển thị khi rê chuột
                    color_discrete_sequence= px.colors.qualitative.Safe, # Màu rời rạc
                    projection="natural earth", # Kiểu chiếu bản đồ
                    title='Bản đồ tuổi thọ trung bình trên thế giới') # Tiêu đề bản đồ

fig.update_layout(
    width=1000,  # Chiều rộng của biểu đồ bằng pixel
    height=500,   # Chiều cao của biểu đồ bằng pixel
    margin = dict(t=40, l=50, r=25, b=25))

fig.show()
```

```python
# Đọc dữ liệu
earthquake = pd.read_csv('data\earthquake.csv')

# Xử lý dữ liệu
data = earthquake.dropna(subset=['mag'])
data = earthquake[data.mag >= 0]

data.head()
```

* **time**: Thời gian xảy ra động đất. Thông thường, cột này chứa dấu thời gian chính xác khi trận động đất bắt đầu.
* **latitude** và **longitude**: Tọa độ địa lý của điểm chính xác nơi động đất xảy ra. Latitude (vĩ độ) và longitude (kinh độ) được đo bằng độ.
* **depth**: Độ sâu của động đất dưới bề mặt Trái Đất, thường được đo bằng kilômét (km).
* **mag**: Độ lớn của động đất. Đây là một chỉ số đo lường tổng năng lượng mà trận động đất giải phóng, thường được đo bằng thang Richter hoặc thang tương tự.
* **place**: Địa điểm xảy ra động đất
* **type** : Loại sự kiện ('earthquake' chỉ đến động đất thông thường, 'ice quake' chỉ các sự kiện địa chấn liên quan đến băng hoặc tuyết)

```python
# Tạo bản đồ Scatter Geo dữ liệu vị trí động đất
fig = px.scatter_geo(data_frame=data, lat='latitude', lon='longitude', 
                     color='mag',
                     hover_name='place', 
                     title='Bản đồ vị trí động đất trên thế giới')


fig.update_layout(
    width=800,  # Chiều rộng của biểu đồ bằng pixel
    height=500,   # Chiều cao của biểu đồ bằng pixel
    margin = dict(t=40, l=50, r=25, b=25))
# Hiển thị biểu đồ
fig.show()
```

#### Trực quan hóa dữ liệu địa lý với mapbox để thể hiện bản đồ chi tiết

```python
top1k = pd.read_csv('data/us-cities-top-1k.csv')
top1k.head()
```

```python
fig = px.scatter_mapbox(top1k, lat='lat', lon='lon',
                       hover_name='City', 
                       hover_data=['State', 'Population'],
                       color_discrete_sequence=['midnightblue'], # màu đơn: midnightblue, darkviolet, deeppink, deepskyblue, salmon,...
                       mapbox_style='open-street-map',
                       zoom=3, height=300)
fig.update_layout(margin={"r":0,"t":0,"l":0,"b":0})
fig.show()
```

```python
# Tùy chỉnh tham số opacity và cho hiển thị 100 thành phố đầu tiên trong tập dữ liệu, màu sắc thay đổi theo Population, 
# Zoom vào khu vực thành phố New York
fig = px.scatter_mapbox(top1k.head(100), lat='lat', lon='lon',
                       hover_name='City', hover_data=['State', 'Population'],
                       size='Population',
                       color_continuous_scale=px.colors.sequential.Plasma, #Dải màu 
                       mapbox_style='open-street-map',
                       opacity=0.9,
                       center={"lat": 40.7128, "lon": -74.0060},
                       zoom=5, height=300)
fig.update_layout(margin={"r":0,"t":0,"l":0,"b":0})
fig.show()
```

#### Hoạt ảnh động - Animation

##### Sử dụng dữ liệu Gapminder từ Plotly Express

```python
# Tải tập dữ liệu Gapminder
gapminder = px.data.gapminder()

# Hiển thị 5 dòng đầu của DataFrame để kiểm tra dữ liệu
gapminder.head()
```

```python
gapminder.year.unique()
```

```python
# Tạo biểu đồ động
fig = px.scatter(gapminder, x="gdpPercap", y="lifeExp", 
                 animation_frame="year", animation_group="country", 
                 size="pop", color="continent", 
                 hover_name="country", log_x=True, size_max=55, 
                 range_x=[100,100000], range_y=[25,90],
                 title="Phát Triển Kinh Tế và Tuổi thọ TB Của Các Quốc Gia Qua Thời Gian")

fig.update_layout(
    width=800,  # Chiều rộng của biểu đồ bằng pixel
    height=400,   # Chiều cao của biểu đồ bằng pixel
    margin = dict(t=40, l=50, r=25, b=25))

# Hiển thị biểu đồ
fig.show()
```

```python
# Vẽ scatter_geo với phép chiếu Natural earth
fig = px.scatter_geo(gapminder, 
                     locations="iso_alpha", 
                     size="pop", 
                     hover_name="country", 
                     color="continent", 
                     title="Phân bố Dân số Của Mỗi Quốc Gia",
                     projection="natural earth",
                    animation_frame="year", 
                 animation_group="country")

fig.update_layout(
    width=800,  # Chiều rộng của biểu đồ bằng pixel
    height=500,   # Chiều cao của biểu đồ bằng pixel
    margin = dict(t=40, l=50, r=25, b=25))
# Hiển thị biểu đồ
fig.show()
```

```python
# Choropleth map với dải màu phân kỳ
fig = px.choropleth(gapminder, 
                    locations="iso_alpha", 
                    color="gdpPercap", 
                    hover_name="country", 
                    color_continuous_scale=px.colors.diverging.Spectral, #Dải màu phân kỳ
                    projection="natural earth",
                    animation_frame="year", 
                     animation_group="country",
                    title="Thu nhập của các quốc gia theo Choropleth Map")

fig.update_layout(
    width=1000,  # Chiều rộng của biểu đồ bằng pixel
    height=500,   # Chiều cao của biểu đồ bằng pixel
    margin = dict(t=40, l=50, r=25, b=25))
# Show the map
fig.show()
```
