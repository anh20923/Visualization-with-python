```python
import plotly.express as px
import pandas as pd
```

```python
# Câu 1: Đọc dữ liệu từ tệp CSV
df = pd.read_csv('Canada.csv',index_col=0)
# Xem 5 dòng đầu
df.head()
```

```python
# Xem 5 dòng cuối
df.tail()
```

```python
# Xem thông tin info()
df.info()
```

```python
# Xem thông tin describe
df.describe()
```

```python
# Xem thông tin describe các cột object
df.describe(include='O')
```

```python
# Câu 2: Vẽ biểu đồ bar chart thể hiện Top 15 Quốc gia nhập cư vào Canada
# Sắp xếp theo cột Total để lấy top 15 quốc gia
df_top15 = df.sort_values('Total',ascending=False).head(15)

# Vẽ biểu đồ cột
fig = px.bar(data_frame=df_top15,y='Country',x='Total',color='Continent',text='Total',height=800)
fig.show()
```

```python
# Câu 3: Vẽ biểu đồ box plot thể hiện Phân bố số người nhập cư vào Canada theo Châu lục
fig = px.box(data_frame=df,x='Continent',y='Total')
fig.show()
```

```python
# Câu 4: Vẽ biểu đồ pie thể hiện Tỷ lệ người nhập cư vào Canada theo Châu lục
# Tính tổng số người nhập cư theo châu lục

# Vẽ biểu đồ pie
fig = px.pie(data_frame=df,names='Continent',values='Total')
fig.show()
```

```python
# Câu 5: Vẽ biểu đồ histogram Phân phối tổng số người nhập cư vào Canada theo Châu lục
fig = px.histogram(data_frame=df,color='Continent',x='Total')
fig.show()
```

##### Hãy sử dụng tính năng tương tác của biểu đồ trên để trả lời cho các câu hỏi sau:

```python
# Câu 6: Vẽ biểu đồ stacked bar tổng số người nhập cư theo mức độ phát triển và châu lục
fig = px.bar(data_frame=df,x='Continent',y='Total',color='DevName',hover_name='Country')
fig.show()
```

##### Hãy sử dụng tính năng tương tác của biểu đồ trên để trả lời cho các câu hỏi sau:
1. Châu lục có số người nhập cư vào Cadana đông nhất? Ít nhất?

```python
# Câu 7: Vẽ treemap về số người nhập cư vào Canada theo Châu lục-Khu vực- Quốc gia
fig = px.treemap(data_frame=df,path=['Continent','Region','Country'],values='Total')
fig.update_layout(width=800,height=600)
fig.show()
```

##### Hãy sử dụng tính năng tương tác của biểu đồ trên để trả lời cho các câu hỏi sau:

```python
# Câu 8: Vẽ biểu đồ treemap về số người nhập cư theo mức độ phát triển và châu lục
fig = px.treemap(data_frame=df,path=['DevName','Continent'],values='Total',height=600)
fig.show()
```

```python
# Câu 9: Vẽ scatter_geo thể hiện Số lượng người nhập cư vào Canada từ các quốc gia
fig = px.scatter_geo(data_frame=df,
                    locations='Country',
                    locationmode='country names',
                    size='Total')
# Hiển thị biểu đồ
fig.show()
```

##### Hãy sử dụng tính năng tương tác của biểu đồ trên để trả lời cho các câu hỏi sau:
1. Số người Việt Nam nhập cư vào Canada?
3. Tìm kiếm và cho biết số người nhập cư vào Canada từ một số quốc gia mà bạn quan tâm.

```python
# Câu 10: Vẽ Choropleth map thể hiện Số lượng người nhập cư vào Canada từ các quốc gia 
fig = px.choropleth(data_frame=df,
                    locations='Country',
                    locationmode='country names',
                    color='Total',
                   projection='natural earth',height=500)
# Hiển thị biểu đồ
fig.show()
```
