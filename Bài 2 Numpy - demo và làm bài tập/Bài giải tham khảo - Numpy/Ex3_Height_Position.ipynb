### Chapter 2 - Exercise 3: Phân tích  chiều cao (height) dựa vào vị trí (position)

**Cho 2 tập tin *heights_2.txt* và _positions.txt_**  
heights_2.txt: chứa dữ liệu về chiều cao của các cầu thủ  
positions.txt: chứa dữ liệu về vị trí thi đấu của các cầu thủ tương ứng  
'GK' (goalkeeper), 'M' (midfield), 'A' (attack) và 'D' (defense)  
(1 cầu thủ sẽ được lưu thông tin chiều cao và vị trí thi đấu ở cùng index)

# Import thư viện

```python
import numpy as np
```

# Phần 1: Đọc dữ liệu vào mảng

```python
# 1. Đọc dữ liệu chiều cao từ tệp "heights_2.txt"
heights = np.genfromtxt("heights_2.txt", delimiter=',', dtype=int)
heights
```

```python
# 2. Đọc dữ liệu vị trí thi đấu từ tệp "heights_2.txt"
positions = np.genfromtxt("positions.txt", delimiter=',', dtype=str)
positions
```

```python
# 3. In ra số phần tử trong mảng chiều cao (số cầu thủ)
print(heights.size)
```

# Phần 2: Truy Xuất Dữ Liệu

```python
# 4. Cho biết chiều cao của cầu thủ đầu tiên trong mảng
heights[0]
```

```python
# 5. Hiển thị vị trí thi đấu của cầu thủ cuối cùng trong mảng
positions[-1]
```

```python
# 6. Cho biết chiều cao của tất cả các cầu thủ tiền đạo "A"
A_heights = heights[positions == "A"]
A_heights
```

```python
# 7. Cho biết chiều cao của các thủ môn "GK"
GK_heights = heights[positions == "GK"]
GK_heights[:10]
```

```python
# 8. Cho biết chiều cao của các tiền vệ "M"
M_heights = heights[positions == "M"]
M_heights[:10]
```

```python
# 9. Cho biết chiều cao của các hậu vệ "D"
D_heights = heights[positions == "D"]
D_heights[:10]
```

```python
# 10. Cho biết vị trí thi đấu của các cầu thủ có chiều cao trên 2m
positions[heights> 200]
```

```python
# 11. Có bao nhiêu thủ môn "GK" cao dưới 1.75m
GK_heights[GK_heights < 175].size
```

```python
# 12. Có bao nhiêu tiền vệ "M" cao trên 1.95m
M_heights[M_heights > 195].size
```

# Phần 3: Thống Kê Dữ Liệu

```python
# 13. Tính chiều cao lớn nhất và nhỏ nhất
max_height = np.max(heights)
min_height = np.min(heights)
print("Chiều cao lớn nhất:", max_height)
print("Chiều cao nhỏ nhất:", min_height)
```

```python
# 14. Cho biết chiều cao của 5 cầu thủ cao nhất và 5 cầu thủ thấp nhất
# 5 cầu thủ thấp nhất
np.sort(heights)[:5]
```

```python
# 5 cầu thủ cao nhất
np.sort(heights)[-5:]
```

```python
# 15. Tính chiều cao trung bình của các cầu thủ
np.mean(heights)
```

```python
# 16. Tính độ lệch chuẩn của chiều cao
np.std(heights)
```

```python
# 17. Tính chiều cao lớn nhất và nhỏ nhất của các thủ môn
max_GK = np.max(GK_heights)
min_GK = np.min(GK_heights)
print("Thủ môn cao nhất:", max_GK)
print("Thủ môn thấp nhất:", min_GK)
```

```python
# 18. Tính chiều cao lớn nhất và nhỏ nhất của các tiền đạo
max_A = np.max(A_heights)
min_A = np.min(A_heights)
print("Tiền đạo cao nhất:", max_A)
print("Tiền đạo thấp nhất:", min_A)
```

```python
# 19. Tính chiều cao lớn nhất và nhỏ nhất của các tiền vệ
max_M = np.max(M_heights)
min_M = np.min(M_heights)
print("Tiền vệ cao nhất:", max_M)
print("Tiền vệ thấp nhất:", min_M)
```

```python
# 20. Tính chiều cao lớn nhất và nhỏ nhất của các hậu vệ
max_D = np.max(D_heights)
min_D = np.min(D_heights)
print("Hậu vệ cao nhất:", max_D)
print("Hậu vệ thấp nhất:", min_D)
```

```python
# 21. Tính số lượng cầu thủ theo vị trí thi đấu
# Số lượng thủ môn
count = np.sum(positions == "GK")
print(f'Goalkeeper : {count} cầu thủ')
```

```python
# Số lượng tiền vệ
count = np.sum(positions == "M")
print(f'Midfield : {count} cầu thủ')
```

```python
names = ['Goalkeeper', 'Midfield', 'Attack', 'Defense']
codes = ['GK', 'M', 'A', 'D']
```

```python
print("Số lượng cầu thủ từng vị trí thi đấu:")
for code, name in zip(codes, names):
    count = np.sum(positions == code)
    print(f'- {name} : {count} cầu thủ')
```

```python
# 22. Cho biết chiều cao trung bình của các 'GK' (goalkeeper), 'M' (midfield), 'A' (attack) và 'D' (defense)
# Tính chiều cao trung bình của các thủ môn GK.
average_height = heights[positions == "GK"].mean()
print(f'Goalkeeper : {round(average_height, 2)}cm')
```

```python
# Tính chiều cao trung bình của các tiền đạo "A".
average_height = heights[positions == "A"].mean()
print(f'Attack : {round(average_height, 2)}cm')
```

```python
# 23. Cho biết chiều cao trung bình của các 'GK' (goalkeeper), 'M' (midfield), 'A' (attack) and 'D' (defense)
print("Chiều cao trung bình của từng vị trí thi đấu:")
for code, name in zip(codes, names):
    average_height = heights[positions == code].mean()
    print(f'- {name}: {round(average_height, 2)}cm')
```
