### Truy xuất phần tử theo điều kiện

```python
arr2
```

```python
arr2 < 70
```

```python
# Hiển thị các phần tử có các giá trị <70
arr2[arr2 < 70]
```

```python
arr2[:, 3] >= 100
```

```python
# Hiển thị giá trị dòng dữ liệu có giá trị cột 3 >100
arr2[arr2[:, 3] >= 100]
```

```python
arr2[2] >= 100
```

```python
# Hiển thị giá trị cột dữ liệu có giá trị dòng 2 >100
arr2[:,arr2[2] >= 100]
```

### Truy xuất phần tử theo nhiều điều kiện

```python
# Hiển thị các phần tử có giá trị từ 100 đến 200
arr2[(arr2 > 100) & (arr2 < 200)] # and
```

```python
# Hiển thị các phần tử có giá trị <= 50 hoặc >= 150
arr2[(arr2 <= 50) | (arr2 >= 150)] # or
```

```python
# Hiển thị các phần tử có giá trị không nhỏ hơn 100 
arr2[~(arr2 < 100)] # not
```

```python
# Hiển thị các dòng có giá trị cột 0 >= 90 và giá trị cột 3 >=100 
arr2[(arr2[:, 0] >= 90) & (arr2[:, 3] >= 100)]
```

### Truy xuất phần tử theo điều kiện với np.where

```python
# Hiển thị các phần tử có giá trị < 70
arr2[idx]
```

## Thống kê trên mảng

```python
# Tạo mảng 2 chiều arr2
arr2 = np.array([[120, 85, 230, 150, 60],
                 [90, 65, 75, 85, 105],
                 [110, 75, 85, 100, 90],
                 [70, 50, 60, 80, 100]])
arr2
```

```python
# Thống kê trên toàn bộ mảng
```

```python
# Tổng tất cả phần tử trong mảng
np.sum(arr2)
```

```python
# phần tử nhỏ nhất trong mảng
np.min(arr2)
```

```python
# phần tử lớn nhất trong mảng
np.max(arr2)
```

```python
# giá trị trung bình của mảng
np.mean(arr2)
```

```python
# giá trị trung vị của mảng
np.median(arr2)
```

```python
# độ lệch chuẩn của mảng
np.std(arr2)
```

```python
# phương sai của mảng
np.var(arr2)
```

```python
# tứ phân vị của mảng
print('Q1 =', np.quantile(arr2, 0.25))
print('Q2 =', np.quantile(arr2, 0.5))
print('Q3 =', np.quantile(arr2, 0.75))
```

```python
# Thống kê các dòng trong cột (axis = 0)
```

```python
arr2
```

```python
print('Tổng các dòng trong cột', np.sum(arr2, axis = 0))

print('Min các dòng trong cột', np.min(arr2, axis = 0))

print('Max các dòng trong cột', np.max(arr2, axis = 0))

print('Mean các dòng trong cột', np.mean(arr2, axis = 0))
```

```python
# Thống kê các cột trong dòng
```

```python
print('Tổng các cột trong dòng', np.sum(arr2, axis = 1))

print('Min các cột trong dòng', np.min(arr2, axis = 1))

print('Max các cột trong dòng', np.max(arr2, axis = 1))

print('Mean các cột trong dòng', np.mean(arr2, axis = 1))
```

```python
# Sắp xếp mảng tăng dần theo cột 0
sorted_indices = np.argsort(arr2[:, 0])
sorted_arr = arr2[sorted_indices, :]
sorted_arr
```

## Kết hợp dữ liệu

```python
arr1
```

```python
arr3 = np.array([95, 80, 120, 60, 150])
```

```python
# Phần tử xuất hiện ở cả 2 mảng arr1, arr3
np.intersect1d(arr1, arr3)
```

```python
# Phần tử chỉ xuất hiện ở arr1 mà không xuất hiện ở arr3
np.setdiff1d(arr1, arr3)
```

```python
# Phần tử chỉ xuất hiện ở arr3 mà không xuất hiện ở arr1
np.setdiff1d(arr3, arr1)
```
