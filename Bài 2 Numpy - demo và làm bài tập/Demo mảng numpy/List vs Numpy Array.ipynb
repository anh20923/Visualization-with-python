## So sánh sự khác nhau giữa List và Numpy


### Import thư viện

```python
import numpy as np
import time
import sys
```

### Về kiểu dữ liệu

```python
# HV tạo mảng chứa thông tin 
my_list = ['Trần Tuấn Anh', 19, 8.5, True]
print(type(my_list), my_list)
```

```python
# HV tạo numpy array từ list thông tin 
my_arr = np.array(my_list)
# chuyển lệnh cho HV, so sánh kết quả HV đã làm
print(type(my_arr), my_arr)
```

```python
# mở rộng bài toán, cho HV quan sát, đoán kiểu dữ liệu
```

### Về dung lượng lưu trữ

```python
# Tạo 1 list số nguyên từ 0-999999
lst_num = list(range(1000000))

# Dung lượng
print(f"Kích thước bộ nhớ của biến lst_num: {sys.getsizeof(lst_num)}")
print(len(lst_num))
```

```python
# Chuyển list thành  mảng NumPy array với cùng dữ liệu
arr_num = np.array(lst_num,dtype=np.int32)

# Dung lượng
print(f"Kích thước bộ nhớ của biến arr_num: {sys.getsizeof(arr_num)} byte")
print(arr_num.size)
```

```python
arr_num.dtype
```

### Thực hiện phép toán

```python
# Nhân 2 cho từng phần tử trong lst_num
lst_cal = [x * 2 for x in lst_num]
lst_cal[:5]
```

```python
# Nhân 2 cho từng phần tử trong arr_num
arr_cal = arr_num * 2
arr_cal[:5]
```

### Tốc độ xử lý

```python
# Bình phương từng phần tử trong lst_num
start_time = time.time()
lst_result = [x ** 2 for x in lst_num]
end_time = time.time()
t0 = end_time - start_time
print(f"Thời gian : {t0:.6f} seconds")
lst_result[:5]
```

```python
# Bình phương từng phần tử trong arr_num
start_time = time.time()
arr_result = arr_num ** 2
end_time = time.time()
t1 = end_time - start_time
print(f"Thời gian: {t1:.6f} seconds")
arr_result[:5]
```
