## Tạo mảng từ các hàm tiện ích
#### Hướng dẫn HV tạo mảng từ các hàm tiện ích

```python
# tạo mảng có 10 phần tử có giá trị 0
arr0 = np.zeros((10,),dtype=int)
arr0
```

```python
# tạo mảng có 10 phần tử có giá trị 1
arr_1 = np.ones((10,),dtype=int)
print(arr_1)
```

```python
# tạo mảng có 10 phần tử có giá trị 4
arr2 = np.full((10,), fill_value=4)
print(arr2)
```

```python
# tạo mảng có 10 phần tử từ 0 đến 9
arr3 = np.arange(10) # arange(0,10,1)
print(arr3)
```

```python
# tạo mảng có 9 phần tử từ 1 đến 9
arr4 = np.arange(1, 10)
print(arr4)
```

```python
# tạo mảng có các phần tử chẵn < 10
arr5 = np.arange(0,10,2)
print(arr5)
```

```python
# tạo mảng có các phần tử 10->1
arr5 = np.arange(10,0,-1)
print(arr5)
```

```python
# linspace(start,stop,num,endpoint)
# endpoint = True => bins= (stop - start)/(num - 1)
# endpoint = False => bins= (stop - start)/(num)
```

```python
# Tạo mảng có 15 phần tử số nguyên cách đều từ 2 đến 30 (bao gồm 30)
arr6 = np.linspace(2, 30, 15,endpoint=True)
print(arr6)
```

```python
arr6 = np.linspace(1, 10, 15,endpoint=False)
print(arr6)
```

```python
# Tạo mảng có 5 phần tử cách đều từ 2 đến 10 (bao gồm 10)
arr7 = np.linspace(2, 10, 5,endpoint=True)
print(arr7)
```

```python
# Tạo mảng có 5 phần tử cách đều từ 2 đến 10 (không bao gồm 10)
arr8 = np.linspace(2, 10, 5, endpoint=False, retstep = True)
print(arr8)
```

```python
# tạo mảng ngẫu nhiên 4 phần tử từ 0 đến 1 (không bao gồm 1)
arr9 = np.random.rand(4)
print(arr9)
```

```python
# tạo mảng ngẫu nhiên 10 phần tử từ 0 đến 5
arr10 = np.random.randint(6,size=10)
print(arr10)
```

### Truy xuất phần tử theo nhiều điều kiện

```python
# Hiển thị các phần tử có giá trị từ 100 đến 200
arr1[(arr1 >= 100) & (arr1 <= 200)] # and
```

```python
# Hiển thị các phần tử có giá trị nhỏ hơn 100 hoặc lớn hơn 200
arr1[(arr1 < 100) | (arr1 > 200)] # or
```

```python
# Hiển thị các phần tử có giá trị không lớn hơn 100
arr1[~(arr1 > 100)]# not
```
